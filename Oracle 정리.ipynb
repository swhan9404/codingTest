```python
""" SQL의 종류
* DQL - 데이터 검색 select
SELECT (distinct 중복항제거) || *(모든) || 컬럼명 as 별칭
FROM 테이블명
WHERE 조건식
ORDER by 컬럼명 (desc || asc)

* DML - 데이터 변경 Insert/Update/Delete
INSERT INTO 테이블명(컬럼명, 컬럼명, ...) values (값, 값, ...)

UPDATE 테이블명 SET 컬럼명=변경값, 컬럼명= 변경값, ... WHERE 조건

DELETE 테이블명 WHERE 조건식
TRUNCATE TALBE 테이블명 #모든 레코드 삭제

* DDL - 데이터 구조정의 Create/ Alter / Drop
CREATE TABLE 테이블이름(
    컬럼명 데이터타입 제약조건들,
    ...
)

ALTER TABLE 테이블명 add ( # 컬럼 추가
    컬럼명 데이터타입 제약조건,
    ...
)
ALTER TABLE 테이블명 drop column 컬럼이름  #컬럼삭제
ALTER TABLE 테이블명 modify 컬럼이름 변경자료형 #자료형 변경
ALTER TABLE 테이블명 rename column 기존컬러명 to 변경컬럼명 #컬럼이름변경

DROP table 테이블명

* DCL - 권한지정 Grant/Revoke
"""

"""
제약조건 종류
1. primary key
2. foreign key  >> CONSTRAINT 제약조건명 REFERENCES 참고테이블명(참고컬럼)
3. unique
4. check
5  default
"""

```
